```python
from bs4 import BeautifulSoup
```

```python
page = open("./Web_data/03. test_first.html","r").read()
soup = BeautifulSoup(page,"html.parser")
print(soup.prettify())
```

```python
# head태그 확인
soup.head
```

```python
# body 태그 확인
soup.body
```

```python
# p 태그 확인
soup.p
```

```python
# 처음 발견한 p 태그만 출력
# find()
soup.find("p")
```

```python
# find_all 모든 p 태그 확인
soup.find_all("p")
```

```python
soup.find("p",class_="inner-text second-item")
```

```python
soup.find("p", {"class" : "outer-text first-item"}).text.strip()
```

```python
# 다중 조건
soup.find("p",{"class": "inner-text first-item","id":"first"})
```

```python
# find_all() : 여러 개의 태그를 반환
# list 형태로 반환

soup.find_all("p")
```

```python
# 특정 태그 확인
soup.find_all(class_="outer-text")
```

## BeautifulSoup 예제 1-1 네이버 금융

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/marketindex/"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

```python
soup.find_all("span","value"), len(soup.find_all("span","value"))
```

```python
soup.find_all("span",{"class" : "value"})[0].text, soup.find_all("span",{"class":"value"})[0].string, soup.find_all("span",{"class":"value"})[0].get_text()
```
