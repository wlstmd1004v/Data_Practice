```python
population = pd.read_excel("./Pop_data/07_population_raw_data.xlsx",header=1)
population.fillna(method="pad",inplace=True)
population
```

```python
population.info()
```

```python
# 컬럼 이름 변경
population.rename(
    columns={
        "행정구역(동읍면)별(1)": "광역시도",
        "행정구역(동읍면)별(2)": "시도",
        "계": "인구수"
    }, inplace=True
)
population.tail()
```

```python
population.head()
```

```python
# 소계 제거
population= population[population["시도"] != "소계"]
population
```

```python
population.is_copy = False

population.rename(
    columns={"항목" : "구분"}, inplace=True
)
```

```python
population.head()
```

```python
population.loc[population["구분"] == "총인구수 (명)", "구분"] = "합계"
population.loc[population["구분"] == "남자인구수 (명)", "구분"] = "남자"
population.loc[population["구분"] == "여자인구수 (명)", "구분"] = "여자"
```

```python
population
```

```python
# 소멸지역을 조사하기 위한 데이터
population["20-39세"] = (
    population["20 - 24세"] + population["25 - 29세"] + population["30 - 34세"] + population["35 - 39세"]
)
population["65세이상"] = (
    population["65 - 69세"] +
    population["70 - 74세"] +
    population["75 - 79세"] +
    population["80 - 84세"] +
    population["85 - 89세"] +
    population["90 - 94세"] +
    population["95 - 99세"]
)
```

```python
population
```

```python
# pivot_table

pop = pd.pivot_table(
    data=population,
    index=["광역시도","시도"],
    columns=["구분"],
    values=["인구수","20-39세","65세이상"]
)
```

```python
pop
```

```python
# 소멸 비율 계산
pop["소멸비율"] = pop["20-39세", "여자"] / (pop["65세이상", "합계"] / 2)
pop.tail()
```

```python
# 소멸위기지역 컬럼 생성
pop["소멸위기지역"] = pop["소멸비율"] < 1.0
pop
```
